```python
import matplotlib.pyplot as plt
import seaborn as sns
import  numpy as np
import matplotlib
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.faker import Faker
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus']=False
with open ("all_genus_name_pie_select_sort.txt") as f:
    genus_name=[]
    genus_num=[]
    explode=[]
    radii=[]
    for i in f:
        i=i.strip().split("\t")
        if i[0]=="Acinetobacter":
            genus_name.append(i[0])
            explode.append(0.2)
            genus_num.append(np.log(6492))
            radii.append(2)
        else:
            genus_name.append(i[0])
            explode.append(0.01)
            genus_num.append(np.log(int(i[1])))
            radii.append(1)
    print(genus_name,genus_num,len(genus_name))
        
```

```python
with open("color1") as f1:
    color=[]
    for i in f1:
        i=i.strip()
        color.append(i)
    print(color)
```

```python
# v = Faker.choose()
# print(v)
legend_opts = opts.LegendOpts(
    orient="vertical",
    pos_top="5%",
    pos_left="5%",
    textstyle_opts=opts.TextStyleOpts(
        font_family="Times New Roman",
        font_size=11,  # 设置字体大小
        color="black"  # 设置字体颜色
    ),
    item_width=10,
    item_height=6,
)
c = (
    Pie()
    .add(
        "",
        [list(z) for z in zip(genus_name, genus_num)],
        radius=["20%", "100%"],
        center=["65%", "60%"],
        rosetype="radius",
        label_opts=opts.LabelOpts(is_show=False),
        
    )
    .set_colors(color)
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie-玫瑰图示例"),
                     legend_opts=legend_opts)#opts.LegendOpts(orient="vertical", pos_top="15%", pos_left="5%"),textstyle_opts=opts.TextStyleOpts(font_family="Times New Roman"))
   # .render("pie_all_genus.html")
)
c.render_notebook()
# make_snapshot(snapshot, c.render(), "pie_all_genus.png")
```
