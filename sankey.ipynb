```python
import plotly.graph_objects as go

# 定义节点标签
node_labels = [
    "源节点1", "源节点2",  # 第一列
    "中间节点1", "中间节点2", "中间节点3",  # 第二列
    "目标节点1", "目标节点2", "目标节点3"  # 第三列
]

# 定义节点颜色
node_colors = [
    "blue", "blue",  # 第一列
    "green", "green", "green",  # 第二列
    "red", "red", "red"  # 第三列
]

# 定义链接
links = dict(
    source = [
        0, 0, 1, 1,  # 第一列到第二列
        2, 3, 4,     # 第二列到第三列
        0, 0, 0, 0,  # 虚拟链接
        1, 1, 1, 1   # 虚拟链接
    ],
    target = [
        2, 3, 3, 4,  # 第一列到第二列
        6, 7, 8,     # 第二列到第三列
        2, 3, 4, 5,  # 虚拟链接
        2, 3, 4, 5   # 虚拟链接
    ],
    value = [
        10, 20, 30, 40,  # 第一列到第二列
        50, 60, 70,      # 第二列到第三列
        0, 0, 0, 0,      # 虚拟链接
        0, 0, 0, 0       # 虚拟链接
    ]
)

# 创建桑基图
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=node_labels,
        color=node_colors
    ),
    link=links
))

# 更新布局和显示图表
fig.update_layout(title_text="三列桑基图示例", font_size=10)
fig.show()
```
