```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # for plot styling
# %matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
from matplotlib import rcParams
```

```python
with open ("all_hotspot_mge_multi_plasmid.txt") as f:
    label=[]
    not_found1=[]
    prophage1=[]
    transpons_puta1=[]
    transpons_pre1=[]
    insert_s1=[]
    satelliter1=[]
    ice1=[]
    ime1=[]
    multiple1=[]
    plasmid1=[]
#     not_found1=[]
#     unoccupy1=[]
#     occupy1=[]
    for i in f:
        i=i.strip().split(",")
#         print(i)
        label.append(i[0])
        not_found1.append((int(i[1])-int(i[2])-int(i[3])-int(i[4])-int(i[5])-int(i[6])-int(i[7])-int(i[8])-int(i[9])-int(i[10]))/int(i[1]))
        prophage1.append(int(i[2])/int(i[1]))
        transpons_puta1.append(int(i[3])/int(i[1]))
        transpons_pre1.append(int(i[4])/int(i[1]))
        insert_s1.append(int(i[5])/int(i[1]))
        ice1.append(int(i[6])/int(i[1]))
        ime1.append(int(i[7])/int(i[1]))
        satelliter1.append(int(i[8])/int(i[1]))
        multiple1.append(int(i[9])/int(i[1]))
        plasmid1.append(int(i[10])/int(i[1]))
#         not_found1.append((int(i[1])-int(i[2])-int(i[3]))/int(i[1]))
#         unoccupy1.append(int(i[2])/int(i[1]))
#         occupy1.append(int(i[3])/int(i[1]))
```

```python
fig = plt.figure(figsize=(8,5))
sns.set_style("white")
# sns.set_style("white")
config = {
    "font.family":'Times New Roman',  # 设置字体类型
#     "font.size": 80,
    "mathtext.fontset":'stix'
}
rcParams.update(config)
plt.bar(label, prophage1, color="#B90328", label="Prophage")#edgecolor= "#B90328",
# 第二根柱子“堆积”在第一根柱子上方，通过'bottom'调整，显示第二种产品的销量
plt.bar(label,transpons_puta1,color="#1E4C9A",label="Putative transpon ",bottom=prophage1)
plt.bar(label,transpons_pre1,color="#4C92C4", label="Transpon ",bottom=np.array(prophage1) + np.array(transpons_puta1))
plt.bar(label,insert_s1,color="#F98D41", label="Insert sequence",bottom=np.array(prophage1) + np.array(transpons_puta1)+np.array(transpons_pre1))
plt.bar(label,ice1,color="#85709B", label="ICE",bottom=np.array(prophage1) + np.array(transpons_puta1)+np.array(transpons_pre1)+np.array(insert_s1))
plt.bar(label,satelliter1,color="#D87093",label="Phage satellite",bottom=np.array(prophage1) + np.array(transpons_puta1)+np.array(transpons_pre1)+np.array(insert_s1)+np.array(ice1))
# plt.bar(label,ime,color="#8FB9BE", label="IME",bottom=np.array(prophage) + np.array(transpons_puta)+np.array(transpons_pre)+np.array(insert_s)+np.array(ice))
plt.bar(label,multiple1,color="#49A9A7",label="Multiple",bottom=np.array(prophage1) + np.array(transpons_puta1)+np.array(transpons_pre1)+np.array(insert_s1)+np.array(ice1)+np.array(satelliter1))
plt.bar(label,plasmid1,color="#FCDF83",label="Plasmid",bottom=np.array(prophage1) + np.array(transpons_puta1)+np.array(transpons_pre1)+np.array(insert_s1)+np.array(ice1)+np.array(satelliter1)+np.array(multiple1))
# 第三根柱子“堆积”在第二根柱子上方，通过'bottom'调整，显示第三种产品的销量
plt.bar(label,not_found1, color="#E7ECF3", 
       bottom=np.array(prophage1) + np.array(transpons_puta1)+np.array(transpons_pre1)+np.array(insert_s1)+np.array(ice1)+np.array(satelliter1)+np.array(multiple1)+np.array(plasmid1), label="None identified")
plt.tick_params(axis='x',length=0)
plt.xlabel('Hotspots',fontsize=12)
plt.ylabel('Defense Island (%)',fontsize=12)
plt.ylim(0,1.01)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.yticks(np.arange(0,1.1,0.1),[f'{i}%' for i in range(0,110,10)])
plt.grid(axis='y',alpha=0.5,ls='--')
plt.legend(loc='lower center',bbox_to_anchor=(0.5,-0.25),ncol=5,frameon=False,handlelength=1, handleheight=1)
# plt.savefig("new_hotspots_mge_is_ice_ime_satelliter_multiple_plasmid.png",bbox_inches = 'tight',dpi=600)
```
